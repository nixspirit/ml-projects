**Task 1:**
Using what you learnt in the scikit-learn decision tree tutorial, use decision trees for classification on the iris dataset, and for regression on the diabetes dataset (both included in ```sklearn.datasets```). Your code should print the accuracy and the confusion matrix for the classification problem, and mean squared error for the regression. Try comparing the results for different maximum tree depths. 

Note: You should split your data 80% training and 20% for testing.



````python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

#Set X a samples times features matrix, Y equal to the targets
X=iris.data 
y=iris.target 

#split to train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#define DT classifier, specificy a maximum tree depth of 3
dt=DecisionTreeClassifier(max_depth=3)

#define training and testing data, fit the classifier
dt.fit(X_train,y_train)

#predict values for test data based on training data
y_pred=dt.predict(X_test)

#print values
print(y_test) # true values
print(y_pred) # predicted values


````

````python
# Evaluate the DT classifier on the iris dataset

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
````

````python
# Now evaluate a regression on the diabetes dataset 
from sklearn.tree import DecisionTreeRegressor

diabetes = datasets.load_diabetes()

#view a description of the dataset 
print(diabetes.DESCR)

#Set X a samples times features matrix, Y equal to the targets
X2=diabetes.data 
y2=diabetes.target 

#split to train and test
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)
````

````python

#define DT classifier, specificy a maximum tree depth of 3
dtr=DecisionTreeRegressor(max_depth=3)

#define training and testing data, fit the classifier
dtr.fit(X2_train,y2_train)

#predict values for test data based on training data
y2_pred=dtr.predict(X2_test)

#print values
#print(y2_test) # true values
#print(y2_pred) # predicted values

# It might be more helpful to plot these outputs

import matplotlib.pyplot as plt
plt.plot(y2_test, color='b', label='true')
plt.plot(y2_pred, color='r', linestyle=':', label='prediction')
plt.legend()


# Evaluate the DT regressor on the diabetes dataset
# As this is a regression problem, we can use R^2 as our metric (see score function)

plt.title('R^2 score: {}'.format(dtr.score(X2_test,y2_test)))


plt.show()
````
